```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan","Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in the US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)

# This can also be written
# fig = plt.figure()
# ax = fig.add_subplot()
```

```python
# Another quick method using object-oriented interface method
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Using plt.show()
## Create the plot
plt.plot(x_axis,y_axis)
plt.show()

# plt.show() should only be used once per session or it could cause problems with output. If this happens 
# restart and clear out put
```

```python
# MATLAB plot editing 

# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Object-Oriented Interface plot editing

# Create the plot and add label
fig, ax= plt.subplots()
ax.plot(x_axis,y_axis, marker="D", color="green", linewidth=2, label="Boston")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")
ax.set_title("Pyber Fare by Month")
ax.set_ylim(0, 45)
ax.grid()
ax.legend()
```

```python
# Create a vertical bar graph

# Create the plot
plt.bar(x_axis,y_axis)
```

```python
# Annotating/ Editing the Bar Graph

# Create the plot
plt.bar(x_axis,y_axis, color="green", label="Boston")
# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fares($)")
# Create a title
plt.title("Pyber Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create a horizontal bar graph

# Create the plot
plt.barh(x_axis,y_axis)
```

```python
# Switching data to opposite axes
plt.barh(y_axis,x_axis)
```

```python
# Inverting the y-axis using gca() method

# Create the plot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Editing/Annotating Horizontal Bar graph using MATLAB approach

plt.barh(x_axis,y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fares($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()

```

```python
# Create bar charts using the object-oriented approach

# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
# Editing / Annotating Horizontal Bar Chart using Object-Oriented Approach
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
# Creating scatter plots using the MATLAB Method

# Create a scatter plot using plt.plot('o')
plt.plot(x_axis,y_axis, 'o')
```

```python
# Create a scatter plot using plt.scatter()
plt.scatter(x_axis,y_axis)
```

```python
# Editing / Annotating a scatter plot using the MATLAB method
plt.scatter(y_axis,x_axis, c='red', label='Chicago')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.xlim(0,45)
plt.title('PyBer Fare by Month')
```

```python
# Creating a Bubble Chart using the MATLAB method
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
# Adding a multipler to all data in the y-axis to adjust the size visually

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
# Using list comprehension to replace the for loop in multiplying the y-axis
plt.scatter(x_axis,y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a scatter plot using the object-oriented interface method

fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Creating a bubble chart using the object-oriented method

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s= y_axis)
```

```python
# Editing / Annotating Bubble charts using the object-oriented method

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='mediumturquoise', s=[i * 5 for i in y_axis], alpha=.5, edgecolor='black', linewidth=2, label='Boston')
plt.gca().invert_yaxis()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()
ax.set_xlim(0,45)
```

```python
# Creating Pie Charts using the MATLAB method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Adding percentages and using the explode() function to enhance our pie chart. Also adding colors and size changes to the pie
plt.subplots(figsize = (8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
        explode=explode_values,
        labels=x_axis,
        colors= colors,
        autopct = '%.1f%%')
```

```python
# Creating pie charts using the object-oriented interface method
fig, ax= plt.subplots()
ax.pie(y_axis, labels=x_axis)
```

```python
# Annotating / Editing our Pie Charts using the Object Oriented Method
fig, ax = plt.subplots(figsize=(8,8))
explode = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,
       labels=x_axis,
       colors=colors,
       explode=explode,
       shadow=True,
       counterclock=False,
       startangle=90,
       autopct='%.1f%%')
```

```python
# Using the Statistics Module to create error bars and find the standard deviation of data

# Get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Create the error bar using the standard deviation
plt.errorbar(x_axis,y_axis, yerr=stdev, capsize=3)
```

```python
# Adding error bars and capsizes using the Object-Oriented method
fig, ax= plt.subplots()
ax.errorbar(x_axis,y_axis, yerr=stdev, capsize = 3)
plt.show()
```

```python
# Adding error bars to a bar chart using the MATLAB approach
plt.bar(x_axis, y_axis, yerr=stdev, capsize= 3)
```

```python
# Adding error bars to a bar chart using the Object-oriented method
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis, yerr=stdev, capsize = 3)
```

```python
# Adjusting major tick marks on a horizontal bar chart using the xticks() function
import numpy as np
plt.barh(x_axis,y_axis)
plt.xticks(np.arange(0, 51, step = 5.0))
plt.gca().invert_yaxis()
```

```python
# Adjusting major tick marks on a horizontal bar chart using the Object-Oriented method
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))
plt.show()
```

```python
# Setting minor ticks on the x-axis of our Horizontal Bar Chart using the Object-Oriented method
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure
fig, ax = plt.subplots(figsize = (8,8))
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))

# Create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
